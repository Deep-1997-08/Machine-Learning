```python
null_value=(data.isnull().sum())
```

```python
a=(null_value.shape)
per_null=[]
for i in (range(a[0])):
    percentage_null=[((null_value.iloc[i]/data.shape[0])*100).round()]
    per_null.append(percentage_null)
    
```

```python
data.shape[0]
```

```python
k=25
for j in (range(a[0])):
    print(data.columns[j],(k-len(data.columns[j]))*" ",int(*per_null[j]),"%")

```
